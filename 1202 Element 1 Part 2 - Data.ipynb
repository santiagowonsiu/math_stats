# Music_Data

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(pd.read_csv('/Users/santiagowon/Dropbox/Santiago/01. Maestria/Advaced Mathematics and Statistics/Element 1 Assignment/data/data.csv'))
df.replace(['', ' ', np.nan, '\u200b'], pd.NA, inplace=True)
numeric_columns = ['tempo', 'beats', 'chroma_stft', 'rmse', 'spectral_centroid', 
                   'spectral_bandwidth', 'rolloff', 'zero_crossing_rate', 'mfcc1',
                   'mfcc2', 'mfcc3', 'mfcc4', 'mfcc5', 'mfcc6', 'mfcc7', 'mfcc8',
                   'mfcc9', 'mfcc10', 'mfcc11', 'mfcc12', 'mfcc13', 'mfcc14',
                   'mfcc15', 'mfcc16', 'mfcc17', 'mfcc18', 'mfcc19', 'mfcc20']


df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
df = df.dropna()
df
```

By inspecting the dataset, we can see that all variables are numeric, suggesting that the dataset contains quantitative information.

#### Scaling Data

```python
# Create a new DataFrame to store the scaled values
df_scaled = df.copy()

# Scale column by multiplying and dividing
df_scaled['tempo'] = df_scaled['tempo'] * 1/10
df_scaled['chroma_stft'] = df_scaled['chroma_stft'] * 100
df_scaled['rmse'] = df_scaled['rmse'] * 100
df_scaled['spectral_centroid'] = df_scaled['spectral_centroid'] * 1/100
df_scaled['spectral_bandwidth'] = df_scaled['spectral_bandwidth'] * 1/100
df_scaled['rolloff'] = df_scaled['rolloff'] * 1/100
df_scaled['zero_crossing_rate'] = df_scaled['zero_crossing_rate'] * 100



print("\nScaled DataFrame:")
df_scaled
```

#### Generating a Leyend with class names - labels

```python
class_names = ["blues", "classical", "country", "disco", "hiphop", "jazz", "metal", "pop", "reggae", "rock"]
class_mapping = {i: class_names[i] for i in range(len(class_names))}

print("Class Mapping:")
for label, class_name in class_mapping.items():
    print(f"Label {label} corresponds to Class {class_name}")
```

### Applying Hinge Loss

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Define your features and target variable
classification_target = 'label'
X_classification = df_scaled.drop(columns=[classification_target, 'filename'])
y_classification = df_scaled[classification_target]

# If the target variable is categorical, use LabelEncoder to convert it to numeric
label_encoder = LabelEncoder()
y_classification_numeric = label_encoder.fit_transform(y_classification)

# Split the data into training and testing sets
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(X_classification, y_classification_numeric, test_size=0.2, random_state=42)

svm_model = SVC(kernel='linear', C=1.0)
svm_model.fit(X_train_cls, y_train_cls)

# Make predictions on the test set
y_pred_svm = svm_model.predict(X_test_cls)

# Evaluate accuracy and print classification report
accuracy_svm = accuracy_score(y_test_cls, y_pred_svm)
print(f'Accuracy (SVM with Hinge Loss): {accuracy_svm}')
print(classification_report(y_test_cls, y_pred_svm))
```

Blues has the highest precision and recall (most accurate).
Rock has the lowest precision and recall.


### Running a Random Forest Classifier model and assess accuracy

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier


classification_target = 'label'
X_classification = df_scaled.drop(columns=[classification_target, 'filename'])
y_classification = df_scaled[classification_target]

# If the target variable is categorical, use LabelEncoder to convert it to numeric
label_encoder = LabelEncoder()
y_classification_numeric = label_encoder.fit_transform(y_classification)

# Split the data into training and testing sets
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(X_classification, y_classification_numeric, test_size=0.2, random_state=42)

# Initialize and fit the Random Forest Classifier
# https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html
random_forest_model = RandomForestClassifier()
random_forest_model.fit(X_train_cls, y_train_cls)

# Make predictions on the test set
y_pred_cls = random_forest_model.predict(X_test_cls)

# Evaluate accuracy and print classification report
accuracy_cls = accuracy_score(y_test_cls, y_pred_cls)
print(f'Accuracy (Random Forest Classifier): {accuracy_cls}')
print(classification_report(y_test_cls, y_pred_cls))
```

We proved same accuracy, while some differences with random forest classifier and hinge loss, there are not major identified differences.

```python
### Analyzing predicted vs true classes classification through an Histogram
```

```python
import matplotlib.pyplot as plt

y_pred_cls = random_forest_model.predict(X_test_cls)

plt.hist(y_pred_cls, bins=len(label_encoder.classes_), alpha=0.7, color='blue', label='Predicted Classes')
plt.hist(y_test_cls, bins=len(label_encoder.classes_), alpha=0.7, color='orange', label='True Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Histogram of Predicted and True Classes')
plt.legend()
plt.show()
```

The distribution and prediction where equal for class 0 and 2

The model predicted more instances in that class compared to the true distribution for class 1,4,6 and 7.

The distribution had more instances than the prediction for classess 3,5,8 and 9

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Generate confusion matrix
conf_matrix = confusion_matrix(y_test_cls, y_pred_cls)

# Get unique labels from y_classification
labels = df_scaled[classification_target].unique()

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Greens', xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

The model achieved high accuracy in predicting Class 6, with 21 correct predictions out of 25 instances. On the other hand, the model struggled the most with Class 9, achieving only 6 correct predictions out of 21 instances.

```python
# Feature Importance 
# https://scikit-learn.org/stable/auto_examples/ensemble/plot_forest_importances.html
# https://inria.github.io/scikit-learn-mooc/python_scripts/dev_features_importance.html

feature_importance = random_forest_model.feature_importances_

# Create a DataFrame to display feature importance
feature_importance_df = pd.DataFrame({'Feature': X_classification.columns, 'Importance': feature_importance})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Plot the feature importance
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df)
plt.title('Feature Importance')
plt.show()
```

chroma_stft has the highest importance, suggesting that it plays a significant role in distinguishing between classes.
Other audio features like rmse, mfcc1, and mfcc4 also have notable importance.
